```python
# function to do the conversion using AudioSegment from pydub. 
# path is used to direct to files/filepath    
def audioConvert(path):
        sound = AudioSegment.from_file(path)
        #set all to mono
        sound = sound.set_channels(1)
        # set to preapproved framerate
        #sound = sound.set_frame_rate()
        # designate ebird code as target directory name(parent) for files
        ebird_code = path.parent.name
        # export to path 
        sound.export((f'/Users/badmin/DSC/birdcall_project/trial/{path.stem}.wav'), format="wav")
```

```python
# function to do the conversion using AudioSegment from pydub. 
# path is used to direct to files/filepath    
def audioConvert(path):
        sound = AudioSegment.from_file(path)
        #set all to mono
        sound = sound.set_channels(1)
        # set to preapproved framerate
        #sound = sound.set_frame_rate()
        # designate ebird code as target directory name(parent) for files
        ebird_code = path.parent.name
        # export to path 
        sound.export((f'/Users/badmin/DSC/birdcall_project/wav_audio/{ebird_code}/{path.stem}.wav'), format="wav")

```

```python

#for (bird)folder in train audio folder
for directory in Path(og_path).iterdir(): 
#create \"new\" path with folder name
    file_paths = list(directory.iterdir())
#change working directory to that folder

# find and convert mp3 to wav
    with Pool(num_workers//2) as PP:
        PP.map(audioConvert, file_paths)
   
```
